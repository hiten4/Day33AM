## Q1 — SVM vs Logistic Regression

Both SVM and Logistic Regression can create linear decision boundaries, but they differ in how they find them.

- **Logistic Regression**: Focuses on minimizing log-loss and outputs probabilities. It uses all data points to learn the boundary.
- **SVM**: Focuses on maximizing the margin between classes and only depends on support vectors (critical points).

**When to use:**
- Use Logistic Regression when probability interpretation is important.
- Use SVM when you want a robust boundary, especially in high-dimensional space.


## Q2 — KNN from Scratch (NumPy)

````python
import numpy as np

def knn_from_scratch(X_train, y_train, X_test, k=3):
    predictions = []

    for test_point in X_test:
        # Compute distances
        distances = np.sqrt(((X_train - test_point) ** 2).sum(axis=1))

        # Get k nearest neighbors
        k_indices = np.argsort(distances)[:k]
        k_labels = y_train[k_indices]

        # Majority vote
        values, counts = np.unique(k_labels, return_counts=True)
        pred = values[np.argmax(counts)]

        predictions.append(pred)

    return np.array(predictions)

print("Function ready to use")
````
